```python
import matplotlib.pyplot as plt

labels = ['Frogs', 'Dogs', 'Cats', 'Birds']

sizes = [15, 30, 45, 10]

explode = (0,0.1,0,0)

fig1, ax1 = plt.subplots()
ax1.pie (sizes, 
         explode=explode, 
         labels=labels, 
         autopct ='%1.1f%%', 
         shadow=True,  
         startangle=90) 
ax1.axis('equal')
```

```python
def flip_coin(): # Creating a loop to make it easier and use one code for the whole process
  coin = (np.random.random() > .5)
  if coin == True: 
    print("Tops")
  if coin == False: 
    print("Bottoms")
```

```python
def flip_coin(): # Code for flipping the coin
  if np.random.random() > 0.5:
    show_img(heads)
  else:
    show_img(tails)
```
